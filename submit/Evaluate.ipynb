```python
import pandas as pd
from tqdm import tqdm
import pickle
import json
from nlgmetricverse import NLGMetricverse, load_metric
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("./answer.pkl", 'rb') as pkl_file:
    answer = pickle.load(pkl_file)
with open("./answer_clip.pkl", 'rb') as pkl_file:
    answer_clip = pickle.load(pkl_file)
with open("./answer_yoclip.pkl", 'rb') as pkl_file:
    answer_yoclip = pickle.load(pkl_file)
```

## CIDEr

```python
yoclip_score=[]
clip_score=[]

scorer = NLGMetricverse(metrics=load_metric("cider"))

for i in tqdm(range(1000)):
    scores = scorer(predictions=[answer_clip[i]], references=[list(answer[i])])
    scores2 = scorer(predictions=[answer_yoclip[i]], references=[list(answer[i])])

    yoclip_score.append(scores2["cider"]["score"])
    clip_score.append(scores["cider"]["score"])
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 将数组转换为numpy数组（如果尚未完成）
yoclip_score = np.array(yoclip_score)
clip_score = np.array(clip_score)

# 创建一个包含两个数组的列表
data = [yoclip_score, clip_score]

# 使用seaborn进行样式设置
sns.set(style="whitegrid")

# 创建箱线图
fig, ax = plt.subplots(figsize=(10, 6))
bplot = ax.boxplot(data, patch_artist=True, labels=['yoclip_score', 'clip_score'])

# 设置箱线图的颜色和填充
colors = ['lightblue', 'lightgreen']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

# 添加网格线
ax.yaxis.grid(True)
ax.set_title('Boxplot of yoclip_score and clip_score', fontsize=16)
ax.set_ylabel('Score', fontsize=14)
ax.set_xlabel('Categories', fontsize=14)

# 添加均值、中位数和标准差的信息
for i, dataset in enumerate(data, start=1):
    mean = np.mean(dataset)
    median = np.median(dataset)
    std_dev = np.std(dataset)
    
    ax.text(i, mean, f'Mean: {mean:.2f}', color='black', ha='center', va='bottom')
    ax.text(i, median, f'Median: {median:.2f}', color='red', ha='center', va='bottom')
    ax.text(i, std_dev, f'Std Dev: {std_dev:.2f}', color='blue', ha='center', va='bottom')

# 显示箱线图
plt.show()
```

```python
import numpy as np

# 将数组转换为numpy数组（如果尚未完成）
yoclip_score = np.array(yoclip_score)
clip_score = np.array(clip_score)

# 打印第一个数组的性质
print("yoclip_score properties:")
print("Mean:", np.mean(yoclip_score))
print("Median:", np.median(yoclip_score))
print("Standard Deviation:", np.std(yoclip_score))
print()

# 打印第二个数组的性质
print("clip_score properties:")
print("Mean:", np.mean(clip_score))
print("Median:", np.median(clip_score))
print("Standard Deviation:", np.std(clip_score))
```
